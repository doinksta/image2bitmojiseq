```python
import numpy as np
import matplotlib.pyplot
import nltk
from nltk.corpus import stopwords
import csv
import re
import gensim

nltk.download('stopwords')
stop_words = stopwords.words('english')
```

```python
captions = []

with open('Train_GCC-training.tsv', 'r') as f:
    reader = csv.reader(f)
    i = 0
    for row in reader:
        if i > 100:
            break
        captions.append(re.split(r'\t+', row[0])[0])
        i += 1
```

```python
captions
```

```python
import gensim.downloader as api
word_vectors = api.load("glove-wiki-gigaword-50")
```

```python

clean_captions = []
for caption in captions:
    tokenized = nltk.word_tokenize(caption)
    new_tokenized = [word_vectors[word] for word in tokenized if word not in stop_words]
    
    # word embeddings
    
    clean_captions.append(new_tokenized)
    
    
    
    
```
